Categorical attributes : pclass can take only 3 different values, survived can take only 2 different values, sex can only take 2 different strings, embarked can only take 3 different values.

```python
df.pclass = df.pclass.astype('category')
df.pclass = df.pclass.cat.rename_categories(["First", "Second","Third"])
df.survived = df.survived.astype('category')
df.survived = df.survived.cat.rename_categories(["Dead", "Survived"])
df.sex = df.sex.astype('category')
df.sex = df.sex.cat.rename_categories(["Female","Male"])
df.embarked = df.embarked.astype('category')
df.embarked = df.embarked.cat.rename_categories(["Cherbourg","Queenstown","Southampton"])
```

Question 4 : For each *travel class*, calculate the proportion of the passengers that survived. Present your results in *pie charts*.

```python

df['bin_survived'] = df.survived.cat.codes
#freq_table1 = pd.crosstab(df.pclass,df.survived)
proportion_table = df.groupby('pclass').mean()
size_of_class = df.groupby('pclass').size()
sum_of_class = df.groupby('pclass').sum()
sum_of_class['number_dead'] = size_of_class - sum_of_class['bin_survived']

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


fig = plt.figure();
fig.suptitle("Proportions of dead and survivors per travel class")
ax1 = plt.subplot(131)
values = sum_of_class.loc['First',['bin_survived','number_dead']]
plt.pie(values,autopct=make_autopct(values))
plt.axis('equal')
plt.title('First class')
plt.legend(['Survived','Dead'])

ax2 = plt.subplot(132)
values = sum_of_class.loc['Second',['bin_survived','number_dead']]
plt.pie(values,autopct=make_autopct(values))
plt.axis('equal')
plt.title('Second class')
plt.legend(['Survived','Dead'])

ax2 = plt.subplot(133)
values = sum_of_class.loc['Third',['bin_survived','number_dead']]
plt.pie(values,autopct=make_autopct(values))
plt.legend(['Survived','Dead'])
plt.axis('equal')
plt.title('Third class')
proportion_table
```

Question 5 : Calculate the proportion of the passengers that survived by *travel class* and *sex*. Present your results in *a single histogram*.

```python
# Create a temporary column with binary values to compute the proportion of survived passenger
# Use the pivot_table method
freq_table_pivot = pd.pivot_table(df,values='bin_survived', index='sex',columns = 'pclass',aggfunc = 'mean')
# Or use groupby method
freq_table_grouped = df.groupby(['sex','pclass']).bin_survived.mean()
# Delete the column added before
#del df['bin_survived']

male_survived = freq_table_grouped.loc['Male']
female_survived = freq_table_grouped.loc['Female']
tmp_df = pd.DataFrame(list(zip(male_survived,female_survived)),columns =['Male','Female'])
ax = tmp_df.plot.bar()
ax.set_xticklabels(('First', 'Second', 'Third'))
ax.set_title('Proportion of passenger that survived by travel class and sex')


del tmp_df
```

```python
df_tmp = pd.DataFrame([freq_table_grouped.loc['Male'].apply(lambda x: x*100), freq_table_grouped.loc['Female'].apply(lambda x: x*100)])
ax2 = df_tmp.plot.bar()
ax2.set_xticklabels(('Male', 'Female'))
ax2.set_title('Proportion of passenger that survived by travel class and sex')
ax2.legend(title="Travel class")

del df_tmp
```

Question 6 : Create 2 equally populated *age categories* and calculate survival proportions by *age category*, *travel class* and *sex*. Present your results in a `DataFrame` with unique index.

```python
#median_age = df['age'].median()
#df.query('age >= @median_age').count()
#df
df_age_sorted = df.sort_values(ascending=True,by=['age'])
df_age_sorted = df_age_sorted.dropna(subset=['age'])
mediane = df_age_sorted.age.size/2
age_category = ['Young']*int(mediane) + ['Old']*int(mediane)
df_age_sorted['age_category'] = age_category
df_age_sorted.age_category = df_age_sorted.age_category.astype('category')
age_class_sex_grouped = df_age_sorted.groupby(['age_category','pclass','sex']).bin_survived.mean()
df_age_sorted = pd.DataFrame(age_class_sex_grouped).apply(lambda x: x*100)
df_age_sorted = df_age_sorted.rename(columns={'bin_survived': 'survivor_proportion'})
print('Index is unique : ' + str(df_age_sorted.index.is_unique))
df_age_sorted
```
